```python
import pandas as pd
```

```python
glaxo=pd.read_csv("D:\\data science\\Priyanka DS\\GLAXO.csv")
glaxo
```

```python
beml=pd.read_csv("D:\\data science\\Priyanka DS\\BEML.csv")
beml
```

```python
glaxo_df=glaxo[['Date','Close']]
beml_df=beml[['Date','Close']]
```

```python
glaxo_df
```

```python
beml_df
```

```python
glaxo_df['Date']=pd.to_datetime(glaxo_df['Date'])
beml_df['Date']=pd.to_datetime(beml_df['Date'])
```

```python
glaxo_df.dtypes
```

```python
beml_df.dtypes
```

```python
glaxo_df=glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date']))
beml_df=beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
```

```python
glaxo_df.drop('Date',axis=1)
```

```python
beml_df.drop('Date',axis=1)
```

```python
glaxo_df['gain']=glaxo_df.Close.pct_change(periods=1)
beml_df['gain']=beml_df.Close.pct_change(periods=1)
```

```python
glaxo_df.drop('Date',axis=1)
```

```python
beml_df.drop('Date',axis=1)
```

```python
glaxo_df.dropna()
beml_df.dropna()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,6))
plt.plot(glaxo_df.gain)
plt.xlabel('Time')
plt.ylabel('Gain')
```

```python
fig=plt.figure(figsize=(10,6))
plt.plot(beml_df.gain)
plt.xlabel('Time')
plt.ylabel('Gain')
```

```python
import seaborn as sn
```

```python
fig=plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sn.distplot(glaxo_df.gain)
plt.xlabel('gain')
plt.ylabel('frequency')
plt.legend()
plt.subplot(1,2,2)
sn.distplot(beml_df.gain)
plt.xlabel('gain')
plt.ylabel('frequency')
plt.legend()
```

```python
print('mean and std dev for galaxo',glaxo_df.gain.mean(),glaxo_df.gain.std())
print('mean and std dev for beml',beml_df.gain.mean(),beml_df.gain.std())
```

```python
from scipy import stats
#find probability for 2% loss or higher for glaxo and beml
print('probability of glaxo for 2%loss',stats.norm.cdf(-0.02,0.000386,0.0133605))
print('probability of beml for 2%loss',stats.norm.cdf(-0.02,0.00027074,0.02643083))
```

```python
#find probability for 2% gain or higher for glaxo and beml
print('probability of glaxo for 2%gain',1-stats.norm.cdf(0.02,0.000386,0.0133605))
print('probability of beml for 2%gain',1-stats.norm.cdf(0.02,0.00027074,0.02643083))
```
